# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız

import random
import time

class Kumanda:
    def __init__(self):
        self.tv_durum = False
        self.ses = 10  # Varsayılan ses seviyesi
        self.kanallar = ["TRT", "Kanal D", "Star Fox", "TV Show"]  # Başlangıç kanalları
        self.acik_kanal = None

    def tv_ac(self):
        if not self.tv_durum:
            print("Televizyon açılıyor...")
            time.sleep(1)
            self.tv_durum = True
            self.acik_kanal = self.kanallar[0]
            print("Televizyon açıldı. Açık kanal:", self.acik_kanal)
        else:
            print("Televizyon zaten açık.")

    def tv_kapat(self):
        if self.tv_durum:
            print("Televizyon kapanıyor...")
            time.sleep(1)
            self.tv_durum = False
            print("Televizyon kapandı.")
        else:
            print("Televizyon zaten kapalı.")

    def ses_ayarları(self):
        if not self.tv_durum:
            print("Önce televizyonu açmalısınız.")
            return
        print(f"Mevcut ses seviyesi: {self.ses}")
        yeni_ses = input("Yeni ses seviyesini girin (0-100): ")
        try:
            yeni_ses = int(yeni_ses)
            if 0 <= yeni_ses <= 100:
                self.ses = yeni_ses
                print("Ses seviyesi güncellendi:", self.ses)
            else:
                print("Lütfen 0-100 arasında bir değer girin.")
        except ValueError:
            print("Geçersiz giriş.")

    def kanal_ekle(self, kanal_ismi=None):
        if kanal_ismi is None:
            kanal_ismi = input("Eklenecek kanal adını girin: ")
        self.kanallar.append(kanal_ismi)
        print(f"{kanal_ismi} kanallara eklendi.")

    def kanal_sayisini_ogren(self):
        print(f"Toplam kanal sayısı: {len(self.kanallar)}")

    def rastgele_kanala_gec(self):
        if not self.tv_durum:
            print("Önce televizyonu açmalısınız.")
            return
        self.acik_kanal = random.choice(self.kanallar)
        print("Rastgele geçilen kanal:", self.acik_kanal)

    def televizyon_bilgileri(self):
        # Durum belirleme
        if self.tv_durum:
            durum = "Açık"
        else:
            durum = "Kapalı"
        print("----- Televizyon Bilgileri -----")
        print(f"Durum: {durum}")
        if self.tv_durum:
            print(f"Açık Kanal: {self.acik_kanal}")
            print(f"Ses Seviyesi: {self.ses}")
        print(f"Kanal Listesi: {', '.join(self.kanallar)}")
        print("-------------------------------")


def menu():
    kumanda = Kumanda()
    print("Kumanda Uygulamasına Hoş Geldiniz!")
    while True:
        print("\nYapmak istediğiniz işlemi seçin:")
        print("q: Çıkış")
        print("1: TV Aç")
        print("2: TV Kapat")
        print("3: Ses Ayarları")
        print("4: Kanal Ekle")
        print("5: Kanal Sayısını Öğrenme")
        print("6: Rastgele Kanala Geçme")
        print("7: Televizyon Bilgileri")

        tercih = input("Seçiminiz: ")
        if tercih.lower() == 'q':
            print("Uygulamadan çıkılıyor...")
            break

        if tercih == "1":
            kumanda.tv_ac()
        elif tercih == "2":
            kumanda.tv_kapat()
        elif tercih == "3":
            kumanda.ses_ayarları()
        elif tercih == "4":
            kumanda.kanal_ekle()
        elif tercih == "5":
            kumanda.kanal_sayisini_ogren()
        elif tercih == "6":
            kumanda.rastgele_kanala_gec()
        elif tercih == "7":
            kumanda.televizyon_bilgileri()
        else:
            print("Geçersiz seçim, tekrar deneyin.")


menu()
```
