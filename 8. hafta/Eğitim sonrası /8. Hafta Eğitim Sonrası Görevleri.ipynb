SÄ±nÄ±flar ise ÅŸu ÅŸekilde: 
* 18, 5 kg/m.'nin altÄ±nda olanlar: ZayÄ±f
* 18.5 â€“ 24, 9 kg/m. arasÄ±nda olanlar: Normal kilolu

```python
# 1.Problemin Ã‡Ã¶zÃ¼mÃ¼nÃ¼ Buraya YazÄ±nÄ±z

  def vki_hesaplama(boy, kilo):
    vki = kilo / (boy ** 2)  
    return vki

def vki_siniflandirma(vki):
    if vki < 18.5:
        return "ZayÄ±f"
    elif 18.5 <= vki < 24.9:
        return "Normal kilolu"
    elif 25 <= vki < 29.9:
        return "Fazla kilolu"
    elif 30 <= vki < 39.9:
        return "Obez"
    else:
        return "Ä°leri derecede obez"

n = int(input("KaÃ§ giriÅŸ yapacaksÄ±nÄ±z? "))
yeni_kayit = []

for i in range(n):
    print(f"\n{i+1}. KULLANICI BÄ°LGÄ°LERÄ°")
    isim = input("Ä°sim: ")
    soyisim = input("Soyisim: ")
    yas = input("YaÅŸ: ")
    boy = float(input("Boy (metre cinsinden): "))
    kilo = float(input("Kilo (kg cinsinden): "))

    vki = vki_hesaplama(boy, kilo)
    sinif = vki_siniflandirma(vki)

    yeni_kayit.append({
        "isim": isim,
        "soyisim": soyisim,
        "yas": yas,
        "boy": boy,
        "kilo": kilo,
        "vki": vki,
        "sinif": sinif
    })

print("\n--- KullanÄ±cÄ± Verileri ---")
for kayit in yeni_kayit:
    print(f"Ä°sim: {kayit['isim']} {kayit['soyisim']}, YaÅŸ: {kayit['yas']}, "
          f"Boy: {kayit['boy']} m, Kilo: {kayit['kilo']} kg, "
          f"VKÄ°: {kayit['vki']:}, SÄ±nÄ±f: {kayit['sinif']}")

with open("kullanicilar.txt", "w", encoding="utf-8") as file:
    file.write("Ä°sim Soyisim | YaÅŸ | Boy (m) | Kilo (kg) | VKÄ° | Kategori\n")
    file.write("-" * 50 + "\n")
    for kayit in yeni_kayit:
        satir = (f"{kayit['isim']} {kayit['soyisim']} | {kayit['yas']} | "
                 f"{kayit['boy']} | {kayit['kilo']} | {kayit['vki']:.2f} | {kayit['sinif']}\n")
        file.write(satir)

print("\nVeriler 'kullanicilar.txt' dosyasÄ±na kaydedildi.")
```
