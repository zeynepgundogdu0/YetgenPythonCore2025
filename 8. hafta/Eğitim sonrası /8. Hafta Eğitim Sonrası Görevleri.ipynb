# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buraya Yazınız

  def vki_hesaplama(boy, kilo):
    vki = kilo / (boy ** 2)  
    return vki

def vki_siniflandirma(vki):
    if vki < 18.5:
        return "Zayıf"
    elif 18.5 <= vki < 24.9:
        return "Normal kilolu"
    elif 25 <= vki < 29.9:
        return "Fazla kilolu"
    elif 30 <= vki < 39.9:
        return "Obez"
    else:
        return "İleri derecede obez"

n = int(input("Kaç giriş yapacaksınız? "))
yeni_kayit = []

for i in range(n):
    print(f"\n{i+1}. KULLANICI BİLGİLERİ")
    isim = input("İsim: ")
    soyisim = input("Soyisim: ")
    yas = input("Yaş: ")
    boy = float(input("Boy (metre cinsinden): "))
    kilo = float(input("Kilo (kg cinsinden): "))

    vki = vki_hesaplama(boy, kilo)
    sinif = vki_siniflandirma(vki)

    yeni_kayit.append({
        "isim": isim,
        "soyisim": soyisim,
        "yas": yas,
        "boy": boy,
        "kilo": kilo,
        "vki": vki,
        "sinif": sinif
    })

print("\n--- Kullanıcı Verileri ---")
for kayit in yeni_kayit:
    print(f"İsim: {kayit['isim']} {kayit['soyisim']}, Yaş: {kayit['yas']}, "
          f"Boy: {kayit['boy']} m, Kilo: {kayit['kilo']} kg, "
          f"VKİ: {kayit['vki']:}, Sınıf: {kayit['sinif']}")

with open("kullanicilar.txt", "w", encoding="utf-8") as file:
    file.write("İsim Soyisim | Yaş | Boy (m) | Kilo (kg) | VKİ | Kategori\n")
    file.write("-" * 50 + "\n")
    for kayit in yeni_kayit:
        satir = (f"{kayit['isim']} {kayit['soyisim']} | {kayit['yas']} | "
                 f"{kayit['boy']} | {kayit['kilo']} | {kayit['vki']:.2f} | {kayit['sinif']}\n")
        file.write(satir)

print("\nVeriler 'kullanicilar.txt' dosyasına kaydedildi.")
```

*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
import datetime

menu = {
    "Çilek": 2,
    "Muz": 5,
    "Kivi": 3,
    "Ahududu": 6,
    "Yaban Mersini": 8,
    "Avokado": 5,
    "Ispanak": 2,
    "Hindistan Cevizi Sütü": 10,
    "Badem Sütü": 6,
    "İnek Sütü": 4,
    "Badem": 7,
}

print("🍹 YetBar - Smoothie Bank'a hoş geldin! ")

while True:
    print("\nYapmak istediğiniz işlemi seçin:")
    print("1. Menüyü görüntüle")
    print("2. Smoothie oluştur ve satın al")
    print("3. Çıkış")

    secim = input("Seçiminiz (1/2/3): ")

    if secim == "1":
        print("\n--- Menü ---")
        for urun, fiyat in menu.items():
            print(f"{urun} - {fiyat} TL")
        print("------------")

    elif secim == "2":
        secilen_urunler = []
        toplam_fiyat = 0

        while True:
            secim = input("Bir malzeme ekleyin (bitirmek için 'tamam' yazın): ")
            if secim == "tamam":
                break

            meyve = False
            for malzeme in menu:
                if secim == malzeme:
                    secilen_urunler.append(malzeme)
                    toplam_fiyat += menu[malzeme]
                    print(malzeme, "eklendi. Toplam:", toplam_fiyat, "TL\n")
                    meyve = True
                    break

            if not meyve:
                print("Geçersiz malzeme adı. Lütfen menüden birini yazın.\n")

        if secilen_urunler:
            tarih = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            print("\n Sipariş Özeti:")
            print("Seçilen ürünler:", ", ".join(secilen_urunler))
            print("Toplam Fiyat:", toplam_fiyat, "TL")
            print("Sipariş Tarihi:", tarih)

            with open("smoothie_siparis_kayitlari.txt", "a", encoding="utf-8") as dosya:
                dosya.write(f"{tarih} - Ürünler: {', '.join(secilen_urunler)} - Toplam: {toplam_fiyat} TL\n")

            print("\n Siparişiniz kaydedildi! Afiyet olsun!\n")

        else:
            print("Hiç malzeme seçilmedi, sipariş oluşturulmadı.\n")

    elif secim == "3":
        print("Görüşmek üzere! ")
        break

    else:
        print("Geçersiz seçim, lütfen 1, 2 veya 3 girin.")


    
```
